```python
# Import Qiskit
import qiskit
from qiskit import QuantumRegister, ClassicalRegister, QuantumCircuit
```

```python
#creating quantum circuit
n_qubits = 5 # number of quantum bits: 3(q0, q1, q2) for encoding the 8 vertices of cube and 2 (q4,q5) for 
             # encoding the possible routes(along the edge).
n_bits = 3   # number of classical bits for storing the measurement
qreg = QuantumRegister(n_qubits) # quantum register with specified number of qubits 
creg = ClassicalRegister(n_bits) # classical register with specified number of bits
qcircuit = QuantumCircuit(qreg, creg) # quantum circuit with specified quantum and classical register 

#Quantum operations 

#apply hadamard to q3 and q4 
qcircuit.h(qreg[3])
qcircuit.h(qreg[4])
#apply cnot with q4 as control and q0 as target
qcircuit.cx(qreg[4], qreg[0])
#apply not to q4
qcircuit.x(qreg[4])
# barrier for better visualization
qcircuit.barrier()
#apply cnot with q4 as control and q1 as target
qcircuit.cx(qreg[4], qreg[1])
#apply cnot with q3 as control and q2 as target
qcircuit.cx(qreg[3], qreg[2])
#apply ccnot with q3 and q4 as controls and q1 as target
qcircuit.ccx(qreg[4], qreg[3], qreg[1])
# barrier for better visualization
qcircuit.barrier()
#apply not to q4
qcircuit.x(qreg[4])
#apply ccnot with q3 and q4 as controls and q0 as target
qcircuit.ccx(qreg[4], qreg[3], qreg[0])
#apply not to q4
qcircuit.x(qreg[4])
#apply ccnot with q3 and q4 as controls and q2 as target
qcircuit.ccx(qreg[4], qreg[3], qreg[2])

# barrier for better visualization
qcircuit.barrier()

# measuring q0, q1 and q2
for i in range (3):
    qcircuit.measure(qreg[i], creg[i])

# draw circuit 
qcircuit.draw(output = 'mpl')
# qcircuit.draw('latex_source', filename='./notes.tex')
```

```python
def qwalk_circuit(input_qubits):
    qreg = QuantumRegister(5) # quantum register with 5 qubits 
    creg = ClassicalRegister(3) # classical register with 3 qubits
    qcircuit = QuantumCircuit(qreg, creg) # quantum circuit with specified quantum and classical register
    
    for iteration, qubit in enumerate(input_qubits):
        if qubit == '1': # to probe the the possible nodes of the starting from a given node on the cube
            qcircuit.x(qreg[2-iteration]) # that is, input qubits for the next step of the walk
    #apply hadamard to q3 and q4 
    qcircuit.h(qreg[3])
    qcircuit.h(qreg[4])
    #apply cnot with q4 as control and q0 as target
    qcircuit.cx(qreg[4], qreg[0])
    #apply not to q4
    qcircuit.x(qreg[4])
    # barrier for better visualization
    qcircuit.barrier()
    #apply cnot with q4 as control and q1 as target
    qcircuit.cx(qreg[4], qreg[1])
    #apply cnot with q3 as control and q2 as target
    qcircuit.cx(qreg[3], qreg[2])
    #apply ccnot with q3 and q4 as controls and q1 as target
    qcircuit.ccx(qreg[4], qreg[3], qreg[1])
    # barrier for better visualization
    qcircuit.barrier()
    #apply not to q4
    qcircuit.x(qreg[4])
    #apply ccnot with q3 and q4 as controls and q0 as target
    qcircuit.ccx(qreg[4], qreg[3], qreg[0])
    #apply not to q4
    qcircuit.x(qreg[4])
    #apply ccnot with q3 and q4 as controls and q2 as target
    qcircuit.ccx(qreg[4], qreg[3], qreg[2])

    # barrier for better visualization
    qcircuit.barrier(range(3))

    # measuring q0, q1 and q2
    for i in range (3):
        qcircuit.measure(qreg[i], creg[i])
    
    return qcircuit    
```

```python
# Adding the transpiler to reduce the circuit to QASM instructions
# supported by the backend
from qiskit import transpile

# Use Aer's qasm_simulator
from qiskit.providers.aer import QasmSimulator

# our simulator
backend = QasmSimulator()

# use histogram for the better viewing of the result
from qiskit.visualization import plot_histogram
```

```python
# Getting all the pitch/notes for running the circuit 30 times
# initial pitch
nodes = ['000']

# Getting a list of 30 pitch
for i in range (30):
    qcircuit = qwalk_circuit(nodes[i])
    # compile the circuit
    qcircuit_compiled = transpile(qcircuit, backend) 
    # Execute the circuit on the qasm simulator.
    # We've set the number of repeats of the circuit
    # to be 1024, which is the default.
    job = backend.run(qcircuit_compiled, shots = 1024)
    # Grab the results from the job.
    result = job.result()
    # Access the counts
    counts = result.get_counts(qcircuit_compiled)
    # We get the keys of the most probable node/pitch from our counts and sotre them in the variable 
    # meas_pitch
    meas_nodes = max(counts, key=counts.get)
    # add the measured notes to our list- notes
    nodes.append(meas_nodes)
    
    # Displaying the counts, hisotogram and circuit for each iteration(each note selected).
    print('=========================','This is the outcome of iteration', i,'==========================')
    print()

    # Displaying how each pitch(note) is produced
    print('This is the final circuit.')
    display(qcircuit.draw(output='mpl'))
    print()    
    
    # View the outcomes using hisogram
    print('This histogram pot gives the better visualization of which pitch is selected.')
    display(plot_histogram(counts))
    print()
```

```python
# Getting all the rythm for running the circuit 30 times
# initial rythm
rythm = ['100']

# Getting a list of 30 notes
for i in range (30):
    qcircuit = qwalk_circuit(rythm[i])
    # compile the circuit
    qcircuit_compiled = transpile(qcircuit, backend) 
    # Execute the circuit on the qasm simulator.
    # We've set the number of repeats of the circuit
    # to be 1024, which is the default.
    job = backend.run(qcircuit_compiled, shots = 1024)
    # Grab the results from the job.
    result = job.result()
    # Access the counts
    counts = result.get_counts(qcircuit_compiled)
    # We get the keys of the most probable rhythm from our counts and sotre them in the variable meas_rythm
    meas_rythm = max(counts, key=counts.get)
    # add the measured rhythm to our list- rhythm
    rythm.append(meas_rythm)
    
    # Displaying the counts, hisotogram and circuit for each iteration(each rythm selected).
    print('=========================','This is the outcome of iteration', i,'===============================')
    print()
    
    # Displaying how each rythm is produced
    print('This is the final circuit.')
    display(qcircuit.draw(output='mpl'))
    print()
    
    # View the outcomes using hisogram
    print('This histogram pot gives the better visualization of which rythm is selected.')
    display(plot_histogram(counts))
    print()
```

```python
# notes_MIDI_num = {'000':65, '001':73, '010':87, '011':98, '100':82, '101':110, '110':65, '111':123}
notes_MIDI_num_dic = {'000':64, '001':68, '010':80, '011':85, '100':90, '101':101, '110':121, '111':128}
rythm_value_dic = {'000':4, '001':3, '010':2, '011':1.5, '100':1, '101':0.75, '110':0.5, '111':0.25}
```

```python
# Now we append MIDI number to all the notes and rythm generated
final_note = []
final_rythm = []
for i in nodes:
    final_note.append(notes_MIDI_num_dic[i]) # all notes generated randomly have now MIDI number(frequency)

for i in rythm:
    final_rythm.append(rythm_value_dic[i])
```

```python
from midiutil import MIDIFile

track    = 0
channel  = 3
time     = 0   # In beats
duration = 1   # In beats, time for which note is played
tempo    = 80  # In BPM
volume   = 100
```

```python
MyMIDI = MIDIFile(1) #single track
MyMIDI.addTempo(track,time, tempo)

for pitch, duration in zip(final_note, final_rythm):
    MyMIDI.addNote(track, channel, pitch, time, duration, volume)
    MyMIDI.addNote(track, channel, pitch-5, time, duration, volume)
    MyMIDI.addNote(track, channel, pitch-7, time, duration, volume) 
    MyMIDI.addNote(track, channel, pitch-12, time, duration, volume)
    MyMIDI.addNote(track, channel, pitch+5, time, duration, volume)
    MyMIDI.addNote(track, channel, pitch+7, time, duration, volume)
    MyMIDI.addNote(track, channel, pitch+12, time, duration, volume)
    time = duration + time + 0.25
    
#output as MIDI file
with open("Quantum_Music.mid", "wb") as output_file:
    MyMIDI.writeFile(output_file)
```
